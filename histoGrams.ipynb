```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("data/img1.jpg")
img_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_RGB)
```

```python
hist_blue = cv2.calcHist([img],channels=[0],mask=None,histSize=[256],ranges=[0,256]) 
```

```python
plt.plot(hist_blue)
```

```python
hist_blue.shape
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    his = cv2.calcHist([img],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(his,color=col)
```
