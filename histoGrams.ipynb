```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("data/img1.jpg")
img_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_RGB)
```

```python
hist_blue = cv2.calcHist([img],channels=[0],mask=None,histSize=[256],ranges=[0,256]) 
```

```python
plt.plot(hist_blue)
```

```python
hist_blue.shape
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    his = cv2.calcHist([img],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(his,color=col)
```

```python
img = cv2.imread("data/img1.jpg")
img_gray = cv2.imread("data/img1.jpg",0)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
his = cv2.calcHist([img_gray],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(his)
```

```python
equi_img = cv2.equalizeHist(img_gray)
```

```python
plt.imshow(equi_img,cmap='gray')
```

```python
his = cv2.calcHist([equi_img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(his)
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
hsv[:,:,2] = cv2.equalizeHist(hsv[:,:,2])
```

```python
equ_color = cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(equ_color)
```
