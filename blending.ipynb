```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Blending images with same shape

```python
img1 = cv2.imread("img1.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("img2.jpg")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
# resizing so that both images are of same size 
img1_re = cv2.resize(img1,(img2.shape[1],img2.shape[0]))
```

```python
img1_re.shape
```

```python
plt.imshow(img1)
```

```python
# blending
blended = cv2.addWeighted(img1_re,0.8,img2,0.5,0)
```

```python
plt.imshow(blended)
```

## blending imgs of different sizes

```python
img1 = cv2.imread("img1.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("img2.jpg")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(250,250))
```

```python
img2.shape
```

```python
img2.dtype
```

```python
plt.imshow(img2)
```

```python
roi = img1[img1.shape[0]-250:img1.shape[0],img1.shape[1]-250:img1.shape[1]]
```

```python
# experimenting with region of interest (roi)
roi = img1[img1.shape[0]-500:img1.shape[0]-250,img1.shape[1]-500:img1.shape[1]-250]
```

```python
plt.imshow(roi)
```

```python
img2_g = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2_g,cmap = 'gray')
```

```python
img2_mask = cv2.bitwise_not(img2_g)
```

```python
plt.imshow(img2_mask,cmap = 'gray')
```

```python
img2_mask.shape
```

```python
final_img2 = cv2.bitwise_or(img2,img2,mask=img2_mask)
```

```python
plt.imshow(final_img2)
```

```python
final = cv2.bitwise_or(roi,final_img2)
```

```python
plt.imshow(final)
```

```python
img1[img1.shape[0]-500:img1.shape[0]-250,img1.shape[1]-500:img1.shape[1]-250] = final
```

```python
plt.imshow(img1)
```
