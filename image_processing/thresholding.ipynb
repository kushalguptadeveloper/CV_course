```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("cross.jpg",0)
```

```python
plt.imshow(img,cmap = 'gray')
```

```python
def show_img(img):
    fig = plt.figure(figsize = (15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
show_img(img)
```

```python
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
show_img(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,8)
```

```python
show_img(th2)
```

```python
final = cv2.addWeighted(th1,0.6,th2,0.4,0)
```

```python
show_img(final)
```
