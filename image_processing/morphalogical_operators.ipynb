```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,text='ABCDE',org=(50,300),fontFace=font,fontScale=5,color=(255,255,255),thickness=20)
    return img
```

```python
def show_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
show_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
result = cv2.erode(img,kernel,iterations=3)
```

```python
show_img(result)
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600))
```

```python
show_img(white_noise)
```

```python
img.max()
```

```python
white_noise = white_noise * 255
```

```python
noise_img = img + white_noise
```

```python
noise_img
```

```python
noise_img.max()
```

```python
show_img(noise_img)
```

```python
noise_img.shape
```

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
```

```python
show_img(opening)
```

```python
black_noise = np.random.randint(low=0,high=2,size=(600,600))
black_noise = black_noise * -255
black_noise_img = img + black_noise
```

```python
black_noise_img
```

```python
black_noise_img[black_noise_img==-255]=0
```

```python
black_noise_img.min()
```

```python
show_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
```

```python
show_img(closing)
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
show_img(gradient)
```
