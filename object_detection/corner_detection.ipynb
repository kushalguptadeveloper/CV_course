```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/Users/kushalgupta/CV_course/data/chess.jpeg")
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
img_gray = np.float32(img_gray)
```

```python
dst = cv2.cornerHarris(src=img_gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst=cv2.dilate(dst,None)
```

```python
img[dst>0.01*dst.max()]=[255,0,0]
```

```python
pl = plt.figure(figsize=(10,10))
ax = pl.add_subplot(111)
ax.imshow(img)
```

# SHi-Tomas

```python
img = cv2.imread("/Users/kushalgupta/CV_course/data/chess.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(img_gray,5,0.01,10)
```

```python
corners
```

```python
corners=np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(img)
```

```python
corners = cv2.goodFeaturesToTrack(img_gray,150,0.01,10)
```

```python
corners=np.int0(corners)
```
