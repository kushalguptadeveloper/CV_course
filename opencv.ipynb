```python
np.shape(fix_img)
```

```python
new_img = cv2.resize(fix_img,(2000,1000))
```

```python
plt.imshow(new_img)
```

```python
# ratio
w_ratio = 0.8
h_ratio = 0.2
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
# flip
new_img = cv2.flip(fix_img,0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,-1)
plt.imshow(new_img)
```

```python
# saving an image
cv2.imwrite("totally_new.jpg",new_img)
```
